Сделать onehot encoding . Тут я заменяю все на индексы внутри одной колонки, так как one hot encoding, создающий плюс тысячу колонок будет дальше мешать

```python
names = []
onehot =[]
for i in list (data['@attribute communityname string ']):
    if i not in names:
        names.append(i)
    onehot.append(names.index(i))

data['@attribute communityname numeric '] = onehot
data = data.drop(columns="@attribute communityname string ")
```

```python
crimes = list(data['@attribute ViolentCrimesPerPop numeric'])
states = list(data['@attribute state numeric '])
d ={}
for state, crime in zip(states, crimes):
    if state in d.keys():
        d[state] += crime
    else:
        d[state] = crime
        
def uniqueish_color(d):
    a = []
    for x in range(len(d)):
        a.append(plt.cm.gist_ncar(np.random.random()))
    return(a)


plt.bar(range(len(d)), list(d.values()), color = uniqueish_color(d))
plt.xticks(range(len(d)), list(d.keys()))
plt.xticks(rotation=90)


plt.show()
```

Построить простую линейную регрессию и оценить score на test и train

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Lasso, Ridge
```

```python
numeric = [x for x in X_test.columns if 'numeric' in x]
```

```python
#### данные здесь должны быть предобработаны на прошлом шаге и все NaN и строковые значения закодированы числами
X_train = X_train[numeric]
X_test = X_test[numeric]

X_train = X_train.fillna(0)
X_test = X_test.fillna(0)
```

```python
%%time
lr = LinearRegression()
lr.fit(X_train,y_train)
```

Постройте график зависимости активных фичей (w>0) от парметра alpha

```python
from tqdm import tqdm
allpoints = []
Y = []
for newalpha in tqdm(np.linspace(0.0001, 0.7, 30)): 
    arr = []
    Y.append(newalpha)
    lass.set_params(alpha = newalpha)
    lass.fit(X_train,y_train)
    for w in lass.coef_:
            if w != 0:
                arr.append(w)
    allpoints.append(arr)
X = []
for ws in allpoints:
    X.append(len(ws))

# дальше посчитать, сколько их было в каждом случае и построить график
```
